```python
import scipy
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,1,10)
points = x**2+1 * np.exp(0.06*np.random.randn(10))

p1 = np.poly1d(np.polyfit(x, points, 1))
p3 = np.poly1d(np.polyfit(x, points, 3))
p7 = np.poly1d(np.polyfit(x, points, 9))


xp = np.linspace(0,1,100)
fig = plt.figure(figsize=(15,5))
# Axis = podgraf, zde jen jeden.
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.plot(xp, p1(xp))
ax1.plot(x, points, '.')
ax2.plot(xp, p3(xp))
ax2.plot(x, points, '.')
ax3.plot(xp, p7(xp))
ax3.plot(x, points, '.')
plt.show()
```
